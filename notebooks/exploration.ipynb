#### Got rid of NaN values by just removing them, removed 'Model' because the value of the column is in 'Make' and dropped 'Engine' b/c I feel the value is in 'Max Power' and 'Max Torque'. Going to multiply length, width, height to get volume so I can drop those three columns and have to see how to fix 'Max Power' and 'Max Torque', if there isn't a large amount of variation in values I might just treat it as a categorical value. 

```python
df['Volume'] = df['Length'] * df['Width'] * df['Height']
df = df.drop(columns=['Length', 'Width', 'Height'])
```

```python
df['Volume'] = (df['Volume']/1000)
## For max power
df['Power_bhp'] = df['Max Power'].str.extract(r'(\d+\.?\d*)\s*bhp').astype(float)
df['Power_rpm'] = df['Max Power'].str.extract(r'@\s*(\d+\.?\d*)\s*rpm').astype(float)
## For max torque
df['Torque_Nm'] = df['Max Torque'].str.extract(r'(\d+\.?\d*)\s*Nm').astype(float)
df['Torque_rpm'] = df['Max Torque'].str.extract(r'@\s*(\d+\.?\d*)\s*rpm').astype(float)
```

```python
df.head()
```

```python
df[['Power_bhp', 'Power_rpm', 'Torque_Nm', 'Torque_rpm']].isnull().sum()
```

```python
df[df['Power_bhp'].isnull()]['Max Power'].unique()
df[df['Torque_Nm'].isnull()]['Max Torque'].unique()
```

```python
df = df.dropna(subset=['Power_bhp', 'Power_rpm', 'Torque_Nm', 'Torque_rpm'])
```

```python
df = df.drop(columns=['Max Power', 'Max Torque'])
```
